### Hilfsmethoden
Zusätzlich zu den Bibliotheken und Datengrundlagen wurden im nachfolgenden Code zwei Funktionen extrahiert, welche häufiger in der nachfolgenden Arbeit verwendet werden. Die eine Funktion zeichnet Kuchendiagramme, während die andere der Kosinus zwischen zwei Datensätzten vergleicht.

Auch dieser Code **muss** ausgeführt werden, damit die Pythondiagramme in den nachfolgenden Kapiteln funktioniert.

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mb
import numpy as np


def get_pie_chart(categories, values, data_source):
    cmap = mb.colormaps['tab20c']
    #cmap = cm.get_cmap('tab20c')  # 'viridis' ist der Name der Colormap
    colors = cmap(np.linspace(0, 1, len(categories)))  # Farben für jede Kategorie generieren
    fig, ax = plt.subplots()
    ax.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=colors)
    ax.axis('equal')  # Gleiches Seitenverhältnis stellt sicher, dass das Pie-Chart als Kreis gezeichnet wird.
    ax.set_facecolor('w')
    plt.text(0.1, -0.1, data_source, horizontalalignment='center',
             verticalalignment='center', transform=ax.transAxes, fontsize=8)

    plt.show()

def cosine_similarity(A, B):
    A_norm = (A - np.mean(A)) / np.std(A)
    B_norm = (B - np.mean(B)) / np.std(B)
    dot_product = np.dot(A_norm, B_norm)

    norm_A = np.linalg.norm(A_norm)
    norm_B = np.linalg.norm(B_norm)

    cosine_sim = dot_product / (norm_A * norm_B)

    return cosine_sim
```

#### Bund/Kantone/Gemeinde Vergleich

Nach der Überprüfung der verschiedenen Einnahmen und Ausgaben auf den einzelnen Ebenen, werden die Daten nun miteinander verglichen. Dabei ist zu beachten das alle Gemeinden und alle Kantone zusammengefasst dargestellt werden. Dieser Vergleich ermöglicht es, Unterschiede und Trends in der finanziellen Entwicklung auf nationaler Ebene zu identifizieren und zu sehen ob eine Ähnlichkeit auf den einzelnen Ebenen besteht.

Zur weiteren Analyse der Daten wird die Cosinus-Ähnlichkeit zwischen den verschiedenen Datensätzen berechnet. Dieses Mass bestimmt, wie ähnlich die Richtungen der jeweiligen Datensatzvektoren sind, unabhängig von ihrer Grösse. Durch die Berechnung der Cosinus-Ähnlichkeit können wir feststellen, wie ähnlich die Trends in den Einnahmen und Ausgaben zwischen den verschiedenen Ebenen oder über die Zeit hinweg sind. Diese Methode ist besonders nützlich, um die Ausrichtung oder Divergenz der finanziellen Trends auf verschiedenen administrativen Ebenen zu quantifizieren.

Die Datensätze bestehen aus effektiven Daten (Durchgezogene Linien) und Prognosen, welche vom Bundesamt für Statistik durchgeführt worden sind (gestrichelte Linie).

```python
import numpy as np
import pandas as pd

df = pd.read_csv(INCOME_SPENDING_CH, delimiter=',', decimal='.')
df_n = df[df['prognose'] == 'n']
df_y = df[df['prognose'] == 'y']

df_y = pd.concat([df_y, df_n.iloc[-1:]])

# Sortiere df_forecast_updated nach 'year', um die Reihenfolge zu gewährleisten
df_y.sort_values('Jahr', inplace=True)

# Erstellen des Plots
plt.figure(figsize=(10, 6))

# Durchgezogene Linien für nicht Prognose Daten
plt.plot(df_n['Jahr'], df_n['Bund_Ertrag'], label='Bund Ertrag', color='blue', linestyle='-')
plt.plot(df_n['Jahr'], df_n['Bund_Ausgaben'], label='Bund Ausgaben', color='red', linestyle='-')
plt.plot(df_n['Jahr'], df_n['Kantone_Ertrag'], label='Kantone Ertrag', color='green', linestyle='-')
plt.plot(df_n['Jahr'], df_n['Kantone_Ausgaben'], label='Kantone Ausgaben', color='orange', linestyle='-')
plt.plot(df_n['Jahr'], df_n['Gemeinden_Ertrag'], label='Gemeinden Ertrag', color='purple', linestyle='-')
plt.plot(df_n['Jahr'], df_n['Gemeinden_Ausgaben'], label='Gemeinden Ausgaben', color='brown', linestyle='-')

# Gestrichelte Linien für Prognose
plt.plot(df_y['Jahr'], df_y['Bund_Ertrag'], color='blue', linestyle='--')
plt.plot(df_y['Jahr'], df_y['Bund_Ausgaben'], color='red', linestyle='--')
plt.plot(df_y['Jahr'], df_y['Kantone_Ertrag'], color='green', linestyle='--')
plt.plot(df_y['Jahr'], df_y['Kantone_Ausgaben'], color='orange', linestyle='--')
plt.plot(df_y['Jahr'], df_y['Gemeinden_Ertrag'], color='purple', linestyle='--')
plt.plot(df_y['Jahr'], df_y['Gemeinden_Ausgaben'], color='brown', linestyle='--')

plt.xlabel('Jahr')
plt.ylabel('Betrag')
plt.title('Finanzen: Ertrag und Ausgaben')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

## Arbeitslosenquote

Aus der Arbeitslosenquote kann man die Prozentsatz der arbeitsfähigen Bevölkerung herauslesen, die aktiv nach Arbeit sucht, aber keine Beschäftigung findet. Sie gibt Aufschluss über den Gesundheitszustand der Wirtschaft; eine hohe Quote kann auf eine schwächelnde Wirtschaft hindeuten, während eine niedrige Quote oft mit wirtschaftlicher Stärke verbunden ist. Sie hilft auch, wirtschaftliche Trends zu erkennen.

In diesem Kapitel vergleichen wir die Wirtschaftsdaten der Schweiz, der USA und der EU, ohne sie saisonal zu bereinigen. Es werden nicht bereinigte Daten verwendet, um die natürlichen Veränderungen direkt zu beobachten und zu verstehen, wie sie sich auf die Wirtschaft auswirken. Wichtig zu verstehen ist, das die Daten monatlich erhoben wurden. Daher wurde zusätzlich bei allen Arbeitslosenquoten Daigramme eine zusätzliche Linie eingeführt, welche immer anfangs Jahr den Wert anzeigt. Somit ist es einfacher ersichtlich wie sich die Quote über die Jahre entwickelt hat.

Ein weiterer Faktor ist die Art und Weise, wie die Schweiz ihre Arbeitslosenquote berechnet: Durch die ausschliessliche Zählung der bei den offiziellen Arbeitsämtern gemeldeten Personen kommt sie auf eine niedrigere Zahl als die ILO und Eurostat, die beide die Schweizer Quote mit 4,2% beziffern. ## Bruttoinlandprodukt (BIP)

Das Bruttoinlandsprodukt (BIP) wird verwendet, um die wirtschaftliche Leistung eines Landes innerhalb eines bestimmten Zeitraums zu messen. Es umfasst den Gesamtwert aller Waren und Dienstleistungen, die im Land produziert wurden, und dient als Indikator für die Größe und Gesundheit der Wirtschaft. Es hilft bei der Bewertung des Wirtschaftswachstums, beim Vergleich der Wirtschaftsleistung verschiedener Länder und bei der Planung wirtschaftspolitischer Massnahmen.

### Vergleich der Schweiz mit der EU und USA
Zum Abschluss des Kapitels BIP wird die Ähnlichkeit der einzelnen Verläufe mithilfe der Cosinus Funktion berechnet:

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Daten laden
df_ch = pd.read_csv(BIP_CH, delimiter=';')
df_eu = pd.read_csv(BIP_EU_UNION_ONLY_TRA, delimiter=';')
df_us = pd.read_csv(BIP_US, delimiter=';')
df_us = df_us[~df_us['year'].astype(str).str.contains('p')]

df_ch['Year'] = df_ch['Year'].astype(int)
df_eu['year'] = df_eu['year'].astype(int)
df_us['year'] = df_us['year'].astype(int)

# Gemeinsame Jahre bestimmen
common_years = set(df_ch['Year']) & set(df_eu['year']) & set(df_us['year'])

if not common_years:
    print("Keine gemeinsamen Jahre gefunden. Überprüfen Sie die Daten.")
else:
    # Daten auf gemeinsame Jahre beschränken
    df_ch = df_ch[df_ch['Year'].isin(common_years)]
    df_eu = df_eu[df_eu['year'].isin(common_years)]
    df_us = df_us[df_us['year'].isin(common_years)]

    # Überprüfen, ob Datenreihen nach Einschränkung leer sind
    if df_ch.empty or df_eu.empty or df_us.empty:
        print("Ein oder mehrere Datensätze sind leer nach der Beschränkung auf gemeinsame Jahre.")
    else:
        # Arrays für Cosine Similarity vorbereiten
        values_ch = df_ch['value'].values.reshape(1, -1)
        values_eu = df_eu['value'].values.reshape(1, -1)
        values_us = df_us['value'].values.reshape(1, -1)

        # Cosine Similarity berechnen
        similarity_ch_eu = cosine_similarity(values_ch, values_eu)[0][0]
        similarity_ch_us = cosine_similarity(values_ch, values_us)[0][0]
        similarity_eu_us = cosine_similarity(values_eu, values_us)[0][0]

        # Ergebnisse ausgeben
        print(f"Cosine Similarity zwischen CH und EU: {similarity_ch_eu}")
        print(f"Cosine Similarity zwischen CH und US: {similarity_ch_us}")
        print(f"Cosine Similarity zwischen EU und US: {similarity_eu_us}")


```

Grundsätzlich lässt sich sagen das der BIP Verlauf der EU, der Schweiz und der USA sehr ähnlich sind (Ähnlichkeit < 99). Das ist über alles gesehen eine grössere Ähnlichkeit als zwischen den Einnahmen und Ausgaben der Kantonen, Gemeinden und dem Bund. Zudem lässt sich sehen, das die Schweiz und die EU ähnlicher sind als der Vergleich zwischen der Schweiz un der USA. Die Ähnlichkeit zwischen der Schweiz und der EU sind zudem ähnlicher als zwischen der EU und den USA.

## Verschuldung

Staatsverschuldung bezieht sich auf die Gesamtsumme der Schulden, die ein Staat angehäuft hat. Das Ein- und Ausnahmedefizit, oft als Haushaltsdefizit bezeichnet, ist die Differenz zwischen den Einnahmen eines Staates und seinen Ausgaben in einem bestimmten Zeitraum. Während die Staatsverschuldung die kumulierten Defizite über die Jahre hinweg darstellt, zeigt das Haushaltsdefizit nur die jährliche finanzielle Lücke. Ein anhaltendes Haushaltsdefizit kann zur Erhöhung der Staatsverschuldung führen, da Staaten sich Geld leihen müssen, um ihre Defizite zu finanzieren.

Für die Schweiz wird der Vergleich zwischen den drei Ebenen Bund, Kantone und Gemeinde gemacht. Zudem sind die Ähnlichkeiten mit dem Cosinus zwischen den einzelnen Werten berechnet.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(DEBT_CH, delimiter=';')

sectors = df['sector'].unique()

plt.figure(figsize=(10, 6))

# Erstellen des Plots
plt.figure(figsize=(10, 6))
sector_vectors = {}

for sector in sectors:
    # Daten für den aktuellen Sektor filtern
    sector_data = df[df['sector'] == sector]
    sector_vectors[sector] = sector_data['value'].values
    # Daten für tatsächliche Werte und Prognosen aufteilen
    actual_data = sector_data[sector_data['forecast'] != 'y']
    forecast_data = sector_data[sector_data['forecast'] == 'y']
    forecast_data = pd.concat([forecast_data, actual_data.iloc[-1:]])
    forecast_data.sort_values('year', inplace=True)

    line, = plt.plot(actual_data['year'], actual_data['value'], linestyle='-', label=sector)
    plt.plot(forecast_data['year'], forecast_data['value'], linestyle='--', color=line.get_color())

similarity_df = pd.DataFrame(index=sectors, columns=sectors)

for sector1 in sectors:
    for sector2 in sectors:
        if len(sector_vectors[sector1]) == len(sector_vectors[sector2]):
            similarity = cosine_similarity(sector_vectors[sector1], sector_vectors[sector2])
            similarity_df.at[sector1, sector2] = similarity


# Diagramm beschriften und Legende anzeigen
plt.title('Wertentwicklung nach Sektor')
plt.xlabel('Jahr')
plt.ylabel('Wert')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Diagramm anzeigen
plt.show()
print(similarity_df)
```

Bei der EU hat sich gezeigt, das die nördlichen Länder grundsätzlich besser abschneiden. Jedoch gab es auch andere Länder welche mehr Einnahmen als Ausgaben hatten oder zumindest keine grosse Diskrepanz zwischen den Zahlen hatten. Hierzu gehören Portugal, aber auch die Baltischen Staaten. Mit Blick auf die gesamte EU sind die Ausgaben jedoch stark grösser als die Einnahmen und es scheint, als würde dieser Unterschied in den nächsten Jahren weiter steigen.

**USA**

Die grosse Überaschung war, das die Einnahmen vor 2001 grösser waren als die Ausgaben. Seit dem war die USA in mehrere Kriege verwickelt und hatte mit der Pandemie zu kämpfen. All dies führte dazu, das die Ausgaben in den letzten Jahren stark gestiegen sind im Gegensatz zu den Einnahmen. Einzig der Budget Control Act hatte kurz dazu geführt, das die Ausgaben gesunken sind.



## Arbeitslosenquote
Die grösste Erkenntnisse ist, das die Schweiz gar nicht so gut dasteht. Die Zahlen scheinen nur besser zu sein, da diese anderst gezählt werden als im Ausland. Zudem kann man in den Zahlen di drei Grossen Krisen, welche die ganze Welt betroffen haben (Dotcom, Finanzkrise, COVID19), alle gut in diesen Zahlen beobachten. Mit Blick auf die EU war die Arbeitslosenquote im Schnitt sehr hoch, jedoch hat sich diese seit 2000 stark verbessert. Aber auch hier sind die Schwankungen stark zu sehen. Die stärksten Schwankungen sind jedoch bei den USA zu sehen. Besonders zur Coronazeit hatte sich diese kurze Zeit auf 12% was sehr hoch ist. Es gilt jedoch zu beachten, das die Amerikaner sehr schnell die Leute temporär feuern können und sobald sich die Wirtschaft erholt, schnell wieder einstellen können.




## Bruttoinlandprodukt

Grundsätzlich ist das BIP bei allen Ländern jedes Jahr gestiegen mit der Ausnahme von einem Jahr, während der Pandemie. Bei der Schweiz, wie auch der EU verändert sich das Bild, wenn man sich das BIP pro Land anschaut, im Gegensatz zum BIP pro Kopf. Von den grössten BIP in der EU: Deutschland, Italien und Frankreich, scheiden beim pro Kopf plötzlich Länder wie Luxembourg und Irland besser ab. Dasselbe bei den Kantonen: Zürich und Bern haben den grössten BIP, aber beim Pro Kopf Vergleich schneiden plötzlich Basel Stadt und Zug besser ab.

Zusätzlich wurde untersucht wie der BIP sich ohne Covid hätte entwickeln können. Hier hat sich bei allen Ländern gezeigt, das es ein Wachstum gegeben hat, im Gegensatz zur aktuellen Senkung. Bei allen drei, also der Schweiz, der EU und den Vereinigten Staaten haben sich diese im nachhinein besser entwickelt, als prognostiziert. In der Schweiz und den USA hat das BIP die Prognose bereits im Jahr 2021 die Prognose ohne Covid Pandemie überholt und in der EU ein Jahr darauf im Jahr 2022.

Zudem wurde festgestellt, dass der Verlauf des Bruttoinlandsprodukts (BIP) der EU, der Schweiz und der USA große Ähnlichkeiten aufweist, mit einer Übereinstimmung von über 97%. Diese Homogenität ist sogar größer als die zwischen den Einnahmen und Ausgaben von Kantonen, Gemeinden und dem Bund. Interessanterweise zeigt sich, dass die Schweiz und die EU eine engere Ähnlichkeit zueinander aufweisen als jeder dieser beiden mit den USA. Dies verdeutlicht, dass die wirtschaftlichen Verläufe in der Schweiz und der EU stärker miteinander korrelieren als mit den wirtschaftlichen Entwicklungen in den USA.



## Verschuldung
In der Schweiz zeigen der Bund und die Kantone im Vergleich zu anderen Staaten eine positive Entwicklung, da sie als einzige ihre Schulden aktiv abbauen. Im Gegensatz dazu steht der Indexvergleich, der ein starkes Wachstum der Schulden in den USA aufzeigt. Auch in der EU ist seit der Eurokrise ein Wachstum der Schulden zu verzeichnen, allerdings nicht so ausgeprägt wie in den USA. Während der Eurokrise hatten nicht nur Griechenland, sondern auch Irland und Portugal erheblich zu kämpfen. Allerdings haben sich Irland und Portugal inzwischen positiv entwickelt. Deutschland und Italien hingegen zeigen ein größeres Schuldenwachstum, das jedoch durch ein ebenfalls größeres BIP abgefedert wird. Im Vergleich zwischen der EU, den Kantonen und den USA wird deutlich, dass unterschiedliche wirtschaftliche Dynamiken und politische Massnahmen zu variierenden Entwicklungen in der Schuldenlage führen.

## Fazit
Grundsätzlich lässt sich sagen, dass die Schweiz im Vergleich sehr gut dasteht. Sie ist zwar nicht Spitzenreiter in allen Kategorien – beispielsweise erzielt Norwegen höhere Einnahmen und die Arbeitslosenquote der Schweiz ist, gemessen wie in anderen Ländern, auch nicht die niedrigste. Trotzdem schneidet die Schweiz insgesamt gut ab und hat sich stabiler entwickelt als viele andere Länder. Besonders in Bezug auf die Verschuldung steht die Schweiz gut da. Über die nächsten Jahre sollte sie diese weiter reduzieren und darauf hinarbeiten, mehr Einnahmen als Ausgaben zu erzielen.

Zudem gab es in den letzten 20 Jahren viele Krisen, die zu einer Zunahme der Schulden führten, die sich zumindest in der Schweiz wieder normalisieren sollten. Die Frage bleibt, wie einzelne Staaten in der EU, insbesondere Italien, Spanien und Deutschland, aber auch die USA, mit der zunehmenden Schuldenlast umgehen werden.
